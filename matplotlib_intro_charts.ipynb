```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Vamos a construir un vector de 41 datos equiespaciados comenzando en cero y terminando en 20.
datos_x = np.linspace(0, 20, 41)
datos_y = datos_x**3
print(datos_x)
print(datos_y)
```

```python
# Graficar x contra y (gráfico de líneas)
plt.plot(datos_x, datos_y)
```

```python
# Grafico con descriptores
plt.plot(datos_x,datos_y,label='X contra Y')
plt.title('Grafico de lineas')
plt.xlabel('Datos de la variable x')
plt.ylabel('Datos de la variable y')
plt.legend()
plt.grid(True)
```

## Tipos de gráficos:

Gráficos especiales para análisis descriptivo de datos: barplots, histogram, scatterplot, boxplot, etc... generalmente son el objetivo de SeaBorn (librería para gráficos estadísticos), se pueden acceder con una menor calidad en MatplotLib

```python
# Scatterplot o diagrama de dispersión
# Grafiquemos datos aleatorios
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x,y)
plt.title('Scatterplot')
```

```python
# Cálculo para representar el área en un gráfico
radio = 16
area = np.pi*(radio*np.random.rand(100))**2

# En gráficos de dispersión se puede representar más de dos variables con diferenciadores
# Creemos una variable
color = np.random.rand(100)

plt.scatter(x,y,s=area,c=color,alpha=0.5)
plt.title('Scatterplot para múltiples variables')
```

```python
# Histogramas
# Creemos una muestra de datos aleatorios
from random import sample # librería para crear muestras aleatorias
data = sample(range(1,1001),100) # Crea una muestra de 100 números aleatorios entre 1 y 1000

plt.hist(data, bins=10)
plt.title('Histograma')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
```

```python
# Diagrama de cajas y bigotes (Box plot)
# Representa la densidad de los datos y los valores atípicos o outliers
datos = [np.random.normal(0,sigma,100) for sigma in range(1,6)]
plt.boxplot(datos,vert=True,patch_artist=True)
plt.title('Diagrama de cajas y bigotes')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
```

```python
# Diagramas de violín
datos1 = [np.random.normal(0,sigma,100) for sigma in range(5,100)]
plt.violinplot(datos1, showmeans=True,showmedians=False)
plt.title('Diagramas de violín')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
```

```python
# Comparación de boxplot y violinplot con programación orientada a objetos
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
# Generemos números aleatorios
datos2=[np.random.normal(0,sigma,100) for sigma in range(6,10)]
# Gráfico de violín
axes[0].violinplot(datos,showmeans=False,showmedians=True)
axes[0].set_title('ViolinPlot')
# Gráfico de cajas
axes[1].boxplot(datos2)
axes[1].set_title('BoxPlot')
# Agreguemos líneas horizontales
for x in axes:
  x.yaxis.grid(True)
  x.set_xticks([y+1 for y in range(len(datos2))])
  x.set_xlabel('XLabel')
  x.set_ylabel('YLabel')

# Agregar los nombres a las divisiones del eje x (ticks, labels)
plt.setp(axes,xticks=[y+1 for y in range(len(datos2))],xticklabels=['F1','F2','F3','F4']);
fig.suptitle('ViolinPlot VS BoxPlot',fontsize=14);
```

```python
# Diagrama de torta PIE
labels='Peras','Mangos','Bananos','Manzanas'
sizes=[7,16,25,34]
explode=(0,.1,0,0) # Extraer una porción de la torta

plt.pie(sizes,explode=explode,labels=labels,shadow=True,autopct='%1.1f%%',startangle=90)
plt.axis('equal'); #Relación de aspecto igual, homogeneidad
```

```python
x1=np.linspace(0,5,5)
y1=x1**3
# Diagramas de ERROR
plt.errorbar(x1,y1,xerr=0.2)
plt.title('Diagrama de error')
plt.grid()
```

```python
x2=1
y2=20
yerror=20*0.1 # 10% de error
plt.errorbar(x2,y2,yerr=yerror,capsize=5)
plt.show()
```
