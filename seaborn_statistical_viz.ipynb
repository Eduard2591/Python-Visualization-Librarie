## Introducción a la visualización con SeaBorn

```python
# Cargar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Librería para la visualización estadística

# https://seaborn.pydata.org/
```

```python
# Cargar datos
datos=pd.read_excel("/content/Datos_Ahorro.xlsx")
```

```python
datos.info()
```

```python
datos.head()
```

```python
# Extraer el vector de nombres de variables
datos.columns
```

```python
# Extraer las categorías de las variables categóricas
print(datos.SEXO.value_counts())
print(datos.CIUDAD.value_counts())
```

```python
# DISTPLOT: Muestra la distribución de una variable
sns.displot(datos.SALARIO);
```

```python
sns.displot(datos.SALARIO,kde=True,bins=25);
# Combinación de Histograma con KDE (Distribución de densidad de probabilidad)
```

```python
# JOINPLOT: Permite emparejar dos displots (Análisis bivariado)
# Con el parámetro kind:Tipo de gráfico a comparar: Sacatter, Reg, Resid, KDE, Hex...etc
```

```python
# Comparar un Histograma con un ScatterPlot
sns.jointplot(x='SALARIO',y='AHORRO',data=datos,kind='scatter') # Por defecto se compara con un histograma
```

```python
# Comparar un Histograma con Hexagonal
sns.jointplot(x='SALARIO',y='AHORRO',data=datos,kind='hex');
```

```python
# Comparar un histograma con scatterplot (kde) kind:'reg'
sns.jointplot(x='SALARIO',y='AHORRO',data=datos,kind='reg');

# Modelos predictivos para análisis de series temporales (Modelos de Box Jenkins - Modelos ARIMA)
```

```python
# PAIRPLOTS: grafica relaciones apareadas (Solo aplica para variables numéricas)
# Para variables categóricas se puede usar un tono de color (hue)
# la diagonal de la matriz de gráficos son los histogramas, y los gráficos apareados son scatterplots
```

```python
sns.pairplot(datos); # Cruce de todas las variables numéricas del DF
```

```python
# Datos categóricos: La diagonal de la matriz son KDE, y los gráficos aparedos son scatter
sns.pairplot(datos,hue='SEXO',palette='inferno'); # palette (colormap)
```

```python
sns.pairplot(datos,hue='CIUDAD',palette='inferno'); # palette (colormap)
```

```python
# KDE: son gráficos que estiman la densidad de los datos...
sns.kdeplot(datos.GASTOS);
```

```python
sns.kdeplot(datos.AHORRO);
```

```python
# GRÁFICOS PARA VARIABLES CATEGÓRICAS
# boxplot
# violinplot
# stripplot
# swarmplot
# barplot
# countplot
```

```python
# BarPlot: se pueden agregar datos categóricos usando una función (por defecto es la media)
sns.barplot(x='SEXO',y='AHORRO',data=datos);
```

```python
# Cambiemos el estimador (min, max, std, median, count...) debe ser un escalar.
sns.barplot(x='SEXO',y='AHORRO',data=datos,estimator=np.min);
```

```python
# BoxPlot: pueden ser usados para mostrar la distribución de las variables categóricas
sns.boxplot(hue='SEXO',y='AHORRO',data=datos,palette='rocket');
```

```python
sns.boxplot(hue='SEXO',y='GASTOS',data=datos,palette='rocket');
```

```python
# Se puede cambiar la orientación del gráfico (orient='h')
sns.barplot(data=datos,palette='rocket',orient='h');
```

```python
# Se pueden ver más variables...
sns.boxplot(x='SEXO',y='AHORRO',hue='CIUDAD',data=datos,palette='rocket');
```

```python
# ViolinPlot
sns.violinplot(x='CIUDAD',y='AHORRO',data=datos,hue='SEXO',palette='inferno');
```

```python
# StripPlot: Muestra un diagrama de dispersión donde la varible es categórica
sns.stripplot(x='CIUDAD',y='AHORRO',data=datos);
```

```python
# StripPlot con más variables categóricas
sns.stripplot(x='CIUDAD',y='SALARIO',data=datos,hue='SEXO',jitter=True,palette='rocket');
```

```python
sns.stripplot(x='CIUDAD',y='GASTOS',data=datos,hue='SEXO',palette='rocket')
```

```python
# dodge:separa las categorías para el scatterplot
```

```python
# SwarmPlot: es muy similar al stripplot, sin embargo, los puntos son más ajustados--- ya no se superponen..
# que se corrije con jitter en los stripplots...
sns.swarmplot(x='CIUDAD',y='GASTOS',data=datos,hue='SEXO',palette='inferno',dodge=True,size=5);
```

```python
# Combinación de gráficos categóricos
sns.swarmplot(x='AHORRO',y='CIUDAD',data=datos,color='black',size=2);
sns.violinplot(x='AHORRO',y='CIUDAD',data=datos,palette='rocket');
```

```python
# GRIDS: Mallas... permiten mapear diferentes tipos de gráficos en filas y columnas
datos2=sns.load_dataset('iris')
datos2
```

```python
datos2.species.value_counts() # Datos balanceados.
```

```python
# PAIRGRID: es un subplot para graficar relaciones bivariadas
sns.pairplot(datos2)
# En la diagonal, se muestran los histogramas de frecuencia para cada variable, y los emparejamientos son scatterplots
```

```python
# Ahora mapeamos la malla anterior
y=sns.pairplot(datos2)
y.map(plt.scatter)
```

```python
# Mapear todo
w=sns.pairplot(datos2)
w.map_upper(plt.scatter); #Diagonal superior de la matriz
# w.map_diag(plt.hist); # Diagonal
w.map_lower(sns.kdeplot); # Diagonal inferior
```

```python
# Una versión más simple es el pairgrid
sns.pairplot(datos2,hue='species',palette='inferno');
```
