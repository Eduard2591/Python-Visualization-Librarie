```python
# Cargar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos
datos = pd.read_excel("/content/Datos_Ahorro.xlsx")

# Obtener informacion general de DF
datos.info()
```

```python
# Viusalizar la estructura del conjunto de datos
datos.head()
```

```python
# Resumen de los estadísticos descriptivos del DF
pd.options.display.float_format = '{:.2f}'.format
datos.describe()
```

```python
# Explorar las categorías de las variables "SEXO" y "CIUDAD"
print(datos.SEXO.value_counts())
print(datos.CIUDAD.value_counts())
```

```python
# Gráfico de líneas
datos.SALARIO.plot(kind='line')
```

```python
# Agregando etiquetas al objeto gráfico
datos.SALARIO.plot(title="Gráfico de líneas por defecto",
                   ylabel="Salario en millones de pesos",
                   xlabel="Empleados",
                   legend="SALARIO",
                   grid=True)
```

```python
# Gráfico de torta o gráfico circular
datos.SEXO.value_counts().plot(kind='pie')
```

```python
datos.SEXO.value_counts().plot(kind="pie",autopct='%1.0f%%')
```

```python
# Gráfico de dispersión
datos.plot(kind='scatter',x='SALARIO',y='AHORRO')
```

```python
# Gráfico de barras
datos.CIUDAD.value_counts().plot(kind='bar')
```

```python
colores=["red","green", "blue"]
datos.CIUDAD.value_counts().plot.bar(title="Gráfico de barras",
                   ylabel="Cantidad de empleados",
                   xlabel="Ciudad",
                   color=colores,
                   grid=True)
```

```python
# Gráfico de histograma
datos.SALARIO.plot(kind='hist')
```

```python
datos.SALARIO.plot.hist(title="Gráfico de barras",
                   ylabel="Cantidad de empleados",
                   xlabel="Ciudad",
                   grid=True, bins=20);
```

```python
# gráficos de cajas y bigotes
datos.plot(kind='box')
```

```python
# Gráfico de densidad (kde)
datos.SALARIO.plot(kind='kde')
```

```python
# Gráfico de densidad (density)
datos.plot(kind='density')
```

```python
# Gráfico de área
datos.plot(kind='area')
```

```python
# Gráfico de hexágonos
datos.plot.hexbin(x='SALARIO',y='AHORRO',gridsize=15)
```
